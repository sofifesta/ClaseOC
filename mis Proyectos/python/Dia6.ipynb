# Día 6: Introducción a Entornos Virtuales y Flask

En esta lección aprenderemos a:
1. Crear y gestionar entornos virtuales en Python.
2. Instalar librerías externas (en este caso, **Flask**).
3. Crear y levantar una página web simple con Flask.

## 1. ¿Qué es un Entorno Virtual?

Un **entorno virtual** es un directorio que contiene una instalación independiente de Python. 
Nos permite trabajar en diferentes proyectos con diferentes dependencias (librerías) sin que interfieran entre sí.

**¿Por qué usarlos?**
- Evita conflictos de versiones entre proyectos.
- Mantiene tu sistema global limpio.
- Facilita compartir el proyecto con otros desarrolladores.

## 2. Paso a Paso: Crear un Entorno Virtual

Para estos pasos, generalmente usamos la **terminal** (CMD, PowerShell o la terminal de tu editor).

### Paso 2.1: Crear el entorno
Abre la terminal en la carpeta de tu proyecto y ejecuta:

```bash
# En Windows, Linux y Mac el comando es el mismo para crearlo
python -m venv nombre_entorno
```

*Nota: Es común llamar al entorno simplemente `env` o `venv`.*

### Paso 2.2: Activar el entorno
Una vez creado, debemos activarlo para que la terminal sepa que queremos usar ese Python y no el global.

**En Windows (PowerShell):**
```powershell
.\nombre_entorno\Scripts\Activate
```

**En Windows (CMD):**
```cmd
nombre_entorno\Scripts\activate.bat
```

**En Mac/Linux:**
```bash
source nombre_entorno/bin/activate
```

Sabrás que está activo porque aparecerá `(nombre_entorno)` al principio de tu línea de comandos.

## 3. Instalando Flask

Con el entorno virtual activado, instalamos Flask usando `pip` (el gestor de paquetes de Python).

```bash
pip install flask
```

## 4. Creando nuestra primera aplicación Web

Vamos a crear una aplicación mínima con Flask. 
Aunque estamos en un Notebook, lo ideal para servidores web es escribir el código en un archivo `.py` (por ejemplo `app.py`).

A continuación te muestro el código que necesitarías:

````python
from flask import Flask

# 1. Inicializamos la aplicación Flask
app = Flask(__name__)

# 2. Definimos una "ruta". 
# Esto le dice a Flask qué hacer cuando alguien entra a la página principal ('/')
@app.route("/")
def home():
    return "<h1>¡Hola Mundo desde Flask!</h1><p>Esta es mi primera página web con Python.</p>"

# 3. Definimos otra ruta para probar
@app.route("/contacto")
def contacto():
    return "<h2>Página de contacto</h2><p>Escríbenos a contacto@ejemplo.com</p>"

# 4. Ejecutamos la aplicación (Solo si este archivo se ejecuta directamente)
# Nota: En un entorno real se evita ejecutar app.run() directamente en producción, 
# pero para practicar está perfecto.
if __name__ == "__main__":
    # debug=True permite que el servidor se reinicie si hacemos cambios en el código
    # IMPORTANTE: No ejecutar esta celda dentro de Jupyter, ya que bloqueará el notebook.
    # Mejor copia este código a un archivo app.py y ejecútalo en la terminal.
    app.run(debug=True, port=5000)
````

## 5. Levantando la página (Ejecutar el servidor)

1.  Crea un archivo nuevo llamado `app.py` en la misma carpeta.
2.  Copia el código de arriba y pégalo en `app.py`.
3.  Ve a tu terminal (asegúrate de que el entorno virtual siga activo).
4.  Ejecuta el programa:
    ```bash
    python app.py
    ```
5.  Verás un mensaje que dice algo como `Running on http://127.0.0.1:5000`.
6.  Abre ese link en tu navegador y verás tu página web.

# Parte 2: Mejorando la estructura (Templates y Static)

Hasta ahora escribimos el código HTML directamente dentro de nuestro archivo Python (`return "<h1>...</h1>"`). Esto funciona para ejemplos pequeños, pero para una web real es **muy mala práctica**.

Lo correcto es separar la lógica (Python) de la presentación (HTML, CSS).

## 6. Separando el HTML: La carpeta `templates`

Flask busca los archivos HTML automáticamente en una carpeta llamada `templates`.

1. Crea una carpeta llamada `templates` al lado de tu `app.py`.
2. Crea un archivo `index.html` dentro de esa carpeta.
3. En tu código de Python, usa la función `render_template` para enviar ese archivo.

## 7. Archivos Estáticos (CSS y JS): La carpeta `static`

Las imágenes, archivos CSS (estilos) y JavaScript se llaman "archivos estáticos". Flask los busca en una carpeta llamada `static`.

1. Crea una carpeta llamada `static` al lado de tu `app.py`.
2. Dentro puedes poner tus estilos (ej. `style.css`) o scripts (`script.js`).
3. Para usarlos en tu HTML, no pones la ruta directa, sino que usas `url_for('static', filename='nombre_archivo')`.

## 8. Estructura de Carpetas Recomendada

Así es como debería verse tu proyecto ahora:

```text
mi_proyecto/
├── venv/                <-- Tu entorno virtual
├── app.py               <-- Tu código Python
├── templates/           <-- Carpeta para HTML
│   └── index.html
└── static/              <-- Carpeta para CSS, JS, Imágenes
    ├── style.css
    └── script.js
```

## 9. Código Actualizado Paso a Paso

### A. El archivo `app.py` actualizado
Observa que importamos `render_template` y cambiamos lo que devuelve la función.

````python
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/")
def home():
    # En lugar de escribir HTML aquí, le decimos a Flask:
    # "Busca index.html en la carpeta templates y devuélvelo"
    return render_template("index.html")

if __name__ == "__main__":
    app.run(debug=True, port=5000)
````

### B. El archivo `templates/index.html`
Este es un archivo HTML normal, pero usamos `{{ ... }}` para usar funciones de Flask como `url_for`.

```html
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <title>Mi Página Flask</title>
    <!-- VINCULAR CSS: Flask generará la ruta correcta a la carpeta static -->
    <link rel="stylesheet" href="{{ url_for('static', filename='style.css') }}">
</head>
<body>
    <div class="container">
        <h1>¡Hola desde HTML separado!</h1>
        <p>Este archivo vive en la carpeta <code>templates</code>.</p>
        <button onclick="saludar()">Haz click aquí</button>
    </div>

    <!-- VINCULAR JS -->
    <script src="{{ url_for('static', filename='script.js') }}"></script>
</body>
</html>
```

### C. El archivo `static/style.css`
Un CSS simple para probar que funciona.

```css
body {
    font-family: Arial, sans-serif;
    background-color: #f0f0f0;
    text-align: center;
    padding: 50px;
}

h1 {
    color: #2c3e50;
}

.container {
    background-color: white;
    padding: 20px;
    border-radius: 10px;
    box-shadow: 0 0 10px rgba(0,0,0,0.1);
    display: inline-block;
}
```

### D. El archivo `static/script.js`
```javascript
function saludar() {
    alert("¡Hola! Este JavaScript fue cargado desde la carpeta static.");
}
```
