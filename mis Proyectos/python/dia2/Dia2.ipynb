## 1. Listas (La mochila de datos)

Imagina que tienes que ir al supermercado. No creas una variable para cada cosa (`item1 = "pan"`, `item2 = "leche"`). Usas una **Lista**.

* Las listas se escriben entre corchetes `[]`.
* El primer elemento siempre está en la posición **0**.

```python
# Creando una lista
supermercado = ["Pan", "Leche", "Huevos", "Galletitas"]

# Imprimir toda la lista
print("Lista completa:", supermercado)

# Acceder a un elemento específico (El índice empieza en 0)
print("El primer item es:", supermercado[0])
print("El tercer item es:", supermercado[2])

# Agregar algo nuevo al final
supermercado.append("Gaseosa")
print("Lista actualizada:", supermercado)
```

### 1.1 Métodos principales de las listas en Python:
* .append(elemento)    - Agrega un elemento al final.
* .extend(iterable)    - Agrega todos los elementos de un iterable al final.
* .insert(índice, obj) - Inserta un objeto en una posición específica.
* .remove(valor)       - Elimina la primera aparición de un valor.
* .pop(índice)         - Elimina y devuelve el elemento en la posición dada (por defecto la última).
* .clear()             - Elimina todos los elementos de la lista.
* .index(valor)        - Devuelve el índice de la primera aparición de un valor.
* .count(valor)        - Cuenta cuántas veces aparece un valor en la lista.
* .sort()              - Ordena los elementos de la lista.
* .reverse()           - Invierte el orden de los elementos.
* .copy()              - Devuelve una copia de la lista.

```python
#crear una lista con 5 elementos
miLista = [1, 2, 3, 4, 5]

#imprimir la lista
print(miLista)
#.append() agrega un elemento al final de la lista
miLista.append(6)
print(miLista)      
#.extend() agrega varios elementos al final de la lista
miLista.extend([7, 8, 9])
print(miLista)          
#.insert() agrega un elemento en una posicion especifica
miLista.insert(0, 0)
print(miLista)      
#.remove() remueve un elemento de la lista
miLista.remove(0)
print(miLista)      
#.pop() remueve el ultimo elemento de la lista
miLista.pop()
print(miLista)      
#.index() devuelve el indice de un elemento
print(miLista.index(1))
#.count() devuelve el numero de veces que aparece un elemento
print(miLista.count(1))
#.sort() ordena la lista
miLista.sort()
print(miLista)      
#.reverse() invierte la lista
miLista.reverse()
print(miLista)      
#.copy() copia la lista
miLista2 = miLista.copy()
print(miLista2)      
#.clear() limpia la lista
miLista.clear()
print(miLista)      
```

## 2. Diccionarios (La agenda)

A veces los números (índices 0, 1, 2) son confusos. Queremos guardar datos con un nombre clave. Para eso usamos **Diccionarios**.

* Se escriben entre llaves `{}`.
* Funcionan como `Clave : Valor`.

```python
# Creando un diccionario para un usuario
usuario = {
    "nombre": "Carlos",
    "edad": 28,
    "es_admin": True
    }

# Accediendo a los datos por su clave (no por número)
print("El usuario se llama:", usuario["nombre"])
print("Su edad es:", usuario["edad"])
```

### 2.1 Metodos de los diccionarios
* `.clear()`: Elimina todos los elementos del diccionario.
* `.copy()`: Devuelve una copia del diccionario.
* `.fromkeys()`: Crea un diccionario a partir de una secuencia de llaves y un valor opcional.
* `.get()`: Devuelve el valor de una llave; si no existe, devuelve `None` o un valor por defecto.
* `.items()`: Devuelve una vista de los pares (llave, valor) del diccionario.
* `.keys()`: Devuelve una vista de las llaves del diccionario.
* `.pop()`: Elimina y devuelve el valor asociado a una llave específica.
* `.popitem()`: Elimina y devuelve el último par (llave, valor) insertado.
* `.setdefault()`: Devuelve el valor de una llave; si no existe, la inserta con un valor por defecto.
* `.update()`: Actualiza el diccionario con elementos de otro diccionario o pares llave-valor.
* `.values()`: Devuelve una vista de los valores del diccionario.



```python
# Diccionario inicial para los ejemplos
usuario = {
    "nombre": "Carlos",
    "edad": 28,
    "es_admin": True
}

# 1. copy(): Crea una copia superficial
copia = usuario.copy()
print(f"copy(): {copia}")

# 2. get(): Obtiene un valor de forma segura
print(f"get('nombre'): {usuario.get('nombre')}")
print(f"get('pais', 'No encontrado'): {usuario.get('pais', 'No encontrado')}")

# 3. keys(): Devuelve una lista (vista) de las llaves
print(f"keys(): {usuario.keys()}")

# 4. values(): Devuelve una lista (vista) de los valores
print(f"values(): {usuario.values()}")

# 5. items(): Devuelve pares clave-valor en tuplas
print(f"items(): {usuario.items()}")

# 6. update(): Actualiza o añade múltiples elementos
usuario.update({"edad": 29, "ciudad": "Madrid"})
print(f"update(): {usuario}")

# 7. setdefault(): Retorna el valor si existe; si no, lo inserta con el valor dado
profesion = usuario.setdefault("profesion", "Ingeniero")
print(f"setdefault(): {profesion} -> {usuario}")

# 8. pop(): Elimina una clave y devuelve su valor
valor_eliminado = usuario.pop("es_admin")
print(f"pop('es_admin'): {valor_eliminado} -> {usuario}")

# 9. popitem(): Elimina y devuelve el último par clave-valor añadido
ultimo = usuario.popitem()
print(f"popitem(): {ultimo} -> {usuario}")

# 10. fromkeys(): Crea un nuevo diccionario desde una secuencia de llaves
nuevas_llaves = ["id", "status", "puntos"]
nuevo_dict = dict.fromkeys(nuevas_llaves, 0)
print(f"fromkeys(): {nuevo_dict}")

# 11. clear(): Vacía el diccionario completamente
usuario.clear()
print(f"clear(): {usuario}")
```

## 3. Bucles (Repeticiones)

En programación, odiamos repetir código. Si quieres imprimir números del 1 al 100, no escribes 100 `print`. Usas un bucle.

### 3.1 El Bucle `for` (Para cada cosa...)
Se usa cuando sabemos cuántas veces queremos repetir algo o para recorrer una lista.

```python
colores = ["Rojo", "Verde", "Azul"]

# Leemos: "Para cada color en la lista de colores..."
for color in colores:
    print("El color actual es:", color)
    
print("--- Fin de la lista ---")
```

```python
# Usando range() para repetir un número de veces
# range(5) genera los números 0, 1, 2, 3, 4
numero=int(input("Cuantas veces repetir?: "))
for numero in range(numero):
    print("Contando:", numero+1)
```

### 3.2 El Bucle `while` (Mientras que...)
Se usa cuando no sabemos cuándo va a terminar, pero sabemos la condición.

**¡Cuidado con los bucles infinitos!**

```python
bateria = 5

while bateria > 0:
    numero=int(input("eliga el numero correcto: "))
    if numero==4:
        print("BIEEEEEEEN")
        bateria=0 # Restamos 1 a la batería
    else:
        print("CASI ......")
```

## 4. El factor Suerte: `random`

Para hacer juegos, necesitamos que la computadora elija cosas al azar. Python tiene una librería llamada `random`.

```python
import random
frutas=['manzana','pera','uva','banana']
# Elegir un número al azar entre 1 y 10
numero_secreto = random.randint(1, 10)
numero_secreto_decimal=round(random.uniform(0,10),random.randint(1,10))
SeleccionaLista=random.choice(frutas)
varias=random.sample(frutas,2)
print("El número secreto es:", numero_secreto,"El número secreto con decimal es:",numero_secreto_decimal)
print(SeleccionaLista)
print(varias)
```
