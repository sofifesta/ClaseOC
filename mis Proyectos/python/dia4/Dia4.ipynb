## 1. Librerías (Herramientas listas)

Python tiene miles de herramientas ya creadas. No reinventemos la rueda. Para usarlas, usamos `import`.

```python
import time  # Librería para manejar el tiempo
import math  # Librería para matemáticas complejas

print("Iniciando cuenta regresiva...")
time.sleep(1) # Pausa de 1 segundo
print("3...")
time.sleep(1)
print("2...")
time.sleep(1)
print("1...")

print("La raíz cuadrada de 49 es:", math.sqrt(49))
```

## 2. Manejo de Errores (`try` / `except`)

¿Qué pasa si le pides un número al usuario y escribe "hola"? El programa explota (Crash). Para evitarlo, usamos bloques de protección.

```python
try:
    # Intentamos hacer algo peligroso
    numero = int(input("Por favor, ingresa un número para dividir 100: "))
    resultado = 100 / numero
    print("El resultado es:", resultado)
    
except ValueError:
    # Esto se ejecuta si el usuario escribe texto
    print("❌ Error: ¡Debes escribir un NÚMERO, no letras!")
    
except ZeroDivisionError:
    # Esto se ejecuta si escribe 0
    print("❌ Error: No se puede dividir por cero.")
    
print("El programa continúa sin explotar.")
```
