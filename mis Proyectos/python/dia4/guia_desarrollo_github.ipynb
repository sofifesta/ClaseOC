# Guía Definitiva: Entorno de Desarrollo, Python y GitHub

## 1. Instalación de Python

### En Windows
1. Ve a la página oficial: [python.org/downloads](https://python.org/downloads/) y descarga la última versión.
2. Abre el instalador descargado.
3. **¡PASO CRÍTICO!** En la primera pantalla, marca la casilla que dice **"Add Python to PATH"** (o "Agregar Python al PATH"). Si no marcas esto, nada funcionará en la terminal.
4. Haz clic en "Install Now" y espera a que termine.

### En Linux (Ubuntu / Debian / Mint)
Abre tu terminal e ingresa los comandos de la celda de abajo:

```python
# Actualizar la lista de paquetes
sudo apt update

# Instalar Python, el gestor de paquetes pip y la herramienta para entornos virtuales
sudo apt install python3 python3-pip python3-venv -y

# Verificar que se instaló correctamente (debería devolver la versión)
python3 --version
```

## 4. Git y GitHub: Instalación y Configuración

### A. Cómo instalar Git

**En Windows:**
1. Ve a [git-scm.com/downloads](https://git-scm.com/downloads) y descarga el instalador.
2. Ejecútalo dándole a **"Siguiente"** a todo (las opciones por defecto son perfectas).
3. Se instalará **Git Bash**, la terminal que debes usar de ahora en adelante.

**En Linux:**
Ejecuta esto en tu terminal:

```python
# Actualizar repositorios e instalar Git
sudo apt update
sudo apt install git -y
```

### B. Crear cuenta y el Token de Seguridad (PAT)

1. Crea tu cuenta en [github.com](https://github.com/) y **verifica tu correo electrónico** desde tu bandeja de entrada.
2. Ve a **Settings** > **Developer settings** > **Personal access tokens** > **Tokens (classic)**.
3. Haz clic en **Generate new token (classic)**.
4. **¡IMPORTANTE!** En la lista de permisos, marca la casilla **`repo`** para darle permisos de lectura y escritura a tus repositorios.
5. Genéralo y **CÓPIALO**. Este código (que empieza con `ghp_...`) será tu contraseña para la terminal.

### C. Configurar Git en tu PC (Se hace solo una vez)

Abre tu terminal y preséntate ante Git. En Linux, también le diremos que memorice tu token para no tener que pegarlo cada vez que subas código:

```python
# Configura tu identidad (Reemplaza con tus datos reales)
git config --global user.name "Tu Nombre o Usuario"
git config --global user.email "tu_correo@ejemplo.com"

# (Opcional - Solo Linux) Le dice a Git que memorice tu Token para el futuro
git config --global credential.helper store
```

## 5. Subir o Descargar Proyectos Inicialmente

Si vas a empezar a trabajar y necesitas **descargar** un proyecto que ya existe en GitHub (te pedirá tu usuario y tu Token al presionar Enter):

```python
# Clonar un repositorio a tu PC
git clone https://github.com/Usuario/Repositorio.git
```

Si en cambio vas a **subir por primera vez** un proyecto local a un repositorio nuevo en GitHub:

1. Crea el repositorio vacío en GitHub (botón "New").
2. Abre la terminal en la carpeta de tu proyecto.
3. Ejecuta estos comandos:

```python
# Iniciar Git, agregar archivos y crear el commit
git init
git add .
git commit -m "Mi primer commit: subiendo el proyecto"

# Renombrar rama y conectar con la nube (REEMPLAZA LA URL)
git branch -M main
git remote add origin https://github.com/TuUsuario/TuRepositorio.git

# Subir los archivos a GitHub (Te pedirá tu usuario y tu Token)
git push -u origin main
```

## 6. El Flujo de Trabajo Diario: Pull -> Código -> Push

El verdadero "día a día". Una vez que tienes el proyecto en tu PC conectado a GitHub, **siempre** debes seguir este orden para evitar conflictos de código:

```python
# 1. ACTUALIZAR TU PC LOCAL (¡Hacer siempre al inicio del día!)
# Trae los últimos cambios de GitHub a tu computadora
git pull origin main

# ... [Aquí es donde programas, modificas y guardas tus archivos] ...

# 2. PREPARAR LOS CAMBIOS
# Agregar todo lo que modificaste a la sala de espera
git add .

# 3. TOMAR LA FOTOGRAFÍA (COMMIT)
# Ponerle un mensaje claro a lo que hiciste
git commit -m "Descripción breve de lo que modificaste hoy"

# 4. SUBIR A LA NUBE (Hacer al terminar la tarea)
# Enviar tu trabajo terminado a GitHub
git push origin main
```

## 7. Bonus: Python, Entornos Virtuales y GitHub

**REGLA DE ORO:** Nunca subas la carpeta de tu entorno virtual (`venv`) a GitHub. Al tener rutas locales y ejecutables del sistema, se rompe al descargarlo en otra computadora. En su lugar, se sube una lista de texto.

**En tu proyecto original:**
1. Crea un archivo `.gitignore` y escribe adentro `venv/` para que Git ignore la carpeta.
2. Genera el archivo con la lista exacta de librerías y versiones que usas:

```python
# Esto guarda todas las librerías instaladas en un archivo de texto
pip freeze > requirements.txt
```

**Al clonar tu proyecto en una PC nueva:**
Como la carpeta `venv` no se subió a GitHub, debes reconstruir el entorno en la nueva computadora y decirle a Python que instale la lista:

```python
# 1. Crear entorno nuevo
python -m venv venv

# 2. Activar entorno 
# En Windows usa: venv\Scripts\activate
# En Linux/macOS usa:
source venv/bin/activate

# 3. Instalar todas las librerías necesarias de golpe
pip install -r requirements.txt
```
