# Día 8: Introducción a CSS (Cascading Style Sheets)

## 1. Introducción Académica y Definición

**CSS (Hojas de Estilo en Cascada)** es un lenguaje de diseño gráfico utilizado para definir y crear la presentación de un documento estructurado escrito en un lenguaje de marcado, como HTML (HyperText Markup Language).

### Conceptos Fundamentales:
*   **Separación de Preocupaciones (Separation of Concerns):** HTML se encarga de la *estructura* y semántica del contenido, mientras que CSS se encarga de la *presentación* visual (colores, tipografía, disposición).
*   **Cascada (Cascading):** Se refiere al algoritmo que determina qué estilos se aplican a un elemento cuando hay múltiples reglas en conflicto. La prioridad se basa en la importancia, la especificidad y el orden de origen.
*   **Herencia:** Ciertas propiedades CSS aplicadas a un elemento padre se transmiten a sus elementos hijos (ej. `font-family`, `color`), mientras que otras no (ej. `border`, `padding`).

### Sintaxis Básica
Una regla CSS consta de un **selector** y un **bloque de declaración**.

```css
selector {
    propiedad: valor;
    propiedad: valor;
}
```

*   **Selector:** Indica el elemento HTML al que se aplicará el estilo (ej. `h1`, `.mi-clase`).
*   **Propiedad:** El aspecto específico que se desea cambiar (ej. `color`, `font-size`).
*   **Valor:** El ajuste que se aplica a la propiedad (ej. `blue`, `16px`).


````python
%%html
<!-- Ejemplo Práctico: Sintaxis Básica -->
<style>
    /* Selector de etiqueta */
    p {
        color: darkblue; /* Propiedad: Valor */
        font-size: 18px;
    }
</style>

<p>Este es un párrafo estilizado con CSS interno.</p>
````

## 2. Métodos de Inclusión de CSS

Existen tres formas de aplicar CSS a un documento HTML, listadas en orden de recomendación (de menor a mayor práctica profesional):

1.  **En línea (Inline):** Usando el atributo `style` dentro de la etiqueta HTML.
    *   *Desventaja:* Difícil de mantener, viola la separación de preocupaciones.
2.  **Interno (Internal):** Usando la etiqueta `<style>` dentro del `<head>` del documento.
    *   *Uso:* Útil para estilos únicos de una sola página.
3.  **Externo (External):** Vinculando un archivo `.css` separado usando la etiqueta `<link>`.
    *   *Ventaja:* Maximiza la reutilización y mantenibilidad. Es el estándar de la industria.

```html
<!-- Ejemplo de vinculación externa -->
<link rel="stylesheet" href="estilos.css">
```


## 3. Selectores CSS y Especificidad

Los selectores son patrones utilizados para seleccionar los elementos que se quieren estilizar. La **Especificidad** es el peso que tiene un selector para determinar qué regla CSS prevalece.

### Tipos de Selectores:

1.  **Selector Universal (`*`):** Selecciona todos los elementos.
2.  **Selector de Tipo (Etiqueta):** Selecciona elementos por su nombre de etiqueta (ej. `div`, `p`).
3.  **Selector de Clase (`.`):** Selecciona elementos con un atributo `class` específico. (ej. `.boton`). Ideal para estilos reutilizables.
4.  **Selector de ID (`#`):** Selecciona un elemento único con un atributo `id` específico. (ej. `#encabezado`). Tiene alta especificidad.
5.  **Selector de Atributo:** Selecciona elementos basados en la presencia o valor de un atributo (ej. `input[type="text"]`).

### Combinadores:
*   **Descendiente (`espacio`):** `div p` (todos los `p` dentro de `div`).
*   **Hijo directo (`>`):** `div > p` (solo los `p` que son hijos directos de `div`).
*   **Hermano adyacente (`+`):** `h1 + p` (el `p` inmediatamente después de `h1`).



````python
%%html
<!-- Ejemplo Práctico: Selectores -->
<style>
    /* Clase */
    .destacado {
        background-color: #f0f0f0;
        padding: 5px;
        border-radius: 4px;
        color: #000;
    }
    /* ID */
    #titulo-unico {
        color: crimson;
        text-transform: uppercase;
    }
    /* Descendiente */
    ul li {
        color: green; /* Todos los li dentro de ul */
    }
</style>

<h3 id="titulo-unico">Título con ID</h3>
<p class="destacado">Este párrafo tiene una clase 'destacado'.</p>
<ul>
    <li>Elemento de lista 1</li>
    <li>Elemento de lista 2</li>
</ul>
````

## 4. El Modelo de Caja (Box Model)

Concepto crucial en CSS. Cada elemento en una página web se considera una caja rectangular.

### Componentes de la Caja (desde adentro hacia afuera):
1.  **Content (Contenido):** El texto o imagen real.
2.  **Padding (Relleno):** Espacio transparente entre el contenido y el borde.
3.  **Border (Borde):** Línea que rodea el padding y el contenido.
4.  **Margin (Margen):** Espacio transparente fuera del borde. Separa los elementos entre sí.

### Box-Sizing
La propiedad `box-sizing` define cómo se calculan el ancho y alto de un elemento.
*   `content-box` (default): `width` incluye solo el contenido. Padding y border se suman al tamaño total.
*   `border-box` (recomendado): `width` incluye contenido, padding y border. Facilita el cálculo de tamaños.

```css
* {
    box-sizing: border-box; /* Buenas práctica estándar */
}
```


````python
%%html
<!-- Ejemplo Práctico: Modelo de Caja -->
<style>
    .caja {
        width: 200px;
        height: 100px;
        padding: 20px;
        border: 5px solid black;
        margin: 20px;
        background-color: lightblue;
        /* box-sizing: content-box; (Por defecto: Ancho total = 200 + 40 + 10 = 250px) */
    }
    .caja-border {
        box-sizing: border-box; /* Ancho total fijo en 200px */
        width: 200px;
        height: 100px;
        padding: 20px;
        border: 5px solid red;
        margin: 20px;
        background-color: lightgreen;
    }
</style>

<div class="caja">Box Sizing: Content Box</div>
<div class="caja-border">Box Sizing: Border Box</div>
````

## 5. Tipografía y Sistemas de Color

### Colores
CSS soporta varios formatos de color:
*   **Nombres clave:** `red`, `blue`, `transparent`.
*   **Hexadecimal:** `#RRGGBB` (ej. `#FF5733`).
*   **RGB/RGBA:** `rgb(255, 0, 0)` o `rgba(255, 0, 0, 0.5)` (con canal alfa/transparencia).
*   **HSL/HSLA:** Hue (Matiz), Saturation (Saturación), Lightness (Luminosidad). Muy intuitivo para diseño.

### Tipografía
Propiedades principales:
*   `font-family`: Define la tipografía (ej. `'Arial', sans-serif`).
*   `font-size`: Tamaño (px, em, rem).
*   `font-weight`: Grosor (400 normal, 700 bold).
*   `line-height`: Espaciado entre líneas.
*   `text-align`: Alineación horizontal (left, center, right, justify).


## 6. Sistemas de Diseño (Layouts) Modernos

Hoy utilizamos **Flexbox** y **Grid**.

### Flexbox (Flexible Box Layout)
Diseñado para estructuras unidimensionales (fila O columna).
*   **Contenedor (`display: flex`):** Habilita el contexto flex.
*   **Ejes:** Eje principal (`main axis`) y eje transversal (`cross axis`).
*   **Propiedades clave:**
    *   `justify-content`: Alineación en el eje principal (ej. `center`, `space-between`).
    *   `align-items`: Alineación en el eje transversal (ej. `center`).
    *   `flex-direction`: Dirección de los ítems (`row`, `column`).

### CSS Grid Layout
Diseñado para estructuras bidimensionales (filas Y columnas simultáneamente).
*   `display: grid`
*   `grid-template-columns`: Define las columnas.
*   `gap`: Espacio entre celdas.


````python
%%html
<!-- Ejemplo Práctico: Flexbox -->
<style>
    .flex-container {
        display: flex;
        justify-content: space-around; /* Distribuye espacio igual alrededor */
        align-items: center; /* Centra verticalmente */
        height: 100px;
        background-color: #eee;
        border: 1px solid #ccc;
    }
    .item {
        width: 50px;
        height: 50px;
        background-color: coral;
        text-align: center;
        line-height: 50px;
        color: white;
    }
</style>

<div class="flex-container">
    <div class="item">1</div>
    <div class="item">2</div>
    <div class="item">3</div>
</div>
````

## 7. Diseño Responsivo y Unidades Relativas

El diseño responsivo (RWD) busca que la web se adapte a cualquier tamaño de pantalla.

### Unidades Relativas vs Absolutas
*   **Absolutas:** `px` (píxeles). Fijas, no escalan bien por defecto.
*   **Relativas:**
    *   `%`: Porcentaje relativo al padre.
    *   `em`: Relativo al tamaño de fuente del elemento padre.
    *   `rem`: Relativo al tamaño de fuente del elemento raíz (`html`). (Muy recomendado para accesibilidad).
    *   `vw` / `vh`: Porcentaje del ancho/alto del viewport (ventana del navegador).

### Media Queries
Permiten aplicar estilos condicionales basados en características del dispositivo, como el ancho.

```css
/* Estilos base (Móvil Primero - Mobile First) */
body { font-size: 16px; }

/* Para pantallas mayores a 768px (Tablets/Desktop) */
@media (min-width: 768px) {
    body { font-size: 18px; }
}
```
