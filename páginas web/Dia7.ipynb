# Introducción a HTML (HyperText Markup Language)

¡Bienvenidos a la clase de hoy! Hoy vamos a adentrarnos en el mundo del desarrollo web aprendiendo su lenguaje fundamental: **HTML**.

### ¿Qué es HTML?

HTML no es un lenguaje de programación como Python. Es un **lenguaje de marcado**.

Para entenderlo mejor, imaginemos una casa o incluso el cuerpo humano:
- **HTML es el esqueleto**: Define la estructura, dónde van las vigas (o los huesos), dónde están las habitaciones (o los órganos). Sin él, todo se caería.
- **CSS (que veremos más adelante) es la decoración y la piel**: Define cómo se ve, los colores, el estilo.
- **JavaScript (también más adelante) son los músculos y el sistema nervioso**: Define el movimiento y la interactividad.

Hoy nos enfocaremos en construir ese esqueleto sólido.

## Sintaxis Básica

En HTML, escribimos utilizando **etiquetas** (tags). Las etiquetas son palabras clave rodeadas por paréntesis angulares `<` y `>`.

La mayoría de las etiquetas tienen una **apertura** y un **cierre**:

```html
<etiqueta>Contenido aquí</etiqueta>
```

- La etiqueta de cierre es igual a la de apertura pero tiene una barra inclinada `/` al principio.
- Lo que está dentro se ve afectado por esa etiqueta.

### ¿Qué es un Elemento?

Es crucial diferenciar entre una "etiqueta" y un "elemento".
- **Etiqueta**: Es solo el comando (ej: `<p>` o `</p>`).
- **Elemento**: Es el conjunto completo: **Etiqueta de apertura** + **Contenido** + **Etiqueta de cierre**.

```text
[    ELEMENTO    ]
<p> Hola Mundo </p>
 |      |       |
Tag  Contenido Tag
Ap.            Cierre
```

### Tipos de Elementos
No todos los elementos se comportan igual. Hay 3 categorías principales que debes conocer:

1. **Elementos de Bloque (Block)**:
   - Intentan ocupar **todo el ancho** disponible.
   - Empiezan siempre en una **nueva línea**.
   - Ejemplos: `<h1>`, `<p>`, `<div>`, `<ul>`.

2. **Elementos de Línea (Inline)**:
   - Ocupan solo el **ancho de su contenido**.
   - **No** empiezan en nueva línea, fluyen con el texto.
   - Ejemplos: `<strong>` (negrita), `<a>` (enlace), `<span>`.

3. **Elementos de Reemplazo (Replaced) / Vacíos**:
   - **Replaced Elements**: Son elementos cuyo contenido es *reemplazado* por un recurso externo (como una imagen), no por texto en el código.
   - **Vacíos (Void)**: No tienen etiqueta de cierre porque no guardan contenido dentro de sí mismos.
   - Ejemplos: `<img>` (imagen), `<input>` (campo de texto), `<br>` (salto de línea), `<hr>` (línea horizontal).

### Atributos
Las etiquetas pueden tener **atributos** que nos dan información extra. Se escriben en la etiqueta de apertura:

```html
<etiqueta atributo="valor">Contenido</etiqueta>
```

### Atributos más comunes en HTML

| Atributo | Descripción |
| :--- | :--- |
| **id** | Identificador único para un elemento en toda la página. |
| **class** | Define una o varias clases para aplicar estilos CSS o seleccionar elementos. |
| **style** | Permite aplicar estilos CSS directamente (inline) al elemento. |
| **href** | Especifica la URL del enlace en etiquetas `<a>`. |
| **src** | Indica la ruta de la fuente para archivos en etiquetas `<img>`, `<script>` o `<iframe>`. |
| **alt** | Texto alternativo que describe una imagen si no se puede cargar (SEO y accesibilidad). |
| **title** | Información adicional que aparece como "tooltip" al pasar el cursor. |
| **target** | Define dónde abrir un enlace (ej: `_blank` para una pestaña nueva). |
| **type** | Especifica el tipo de control en `<input>` (ej: `text`, `password`, `checkbox`). |
| **placeholder** | Texto de sugerencia que aparece dentro de un `<input>` antes de escribir. |
| **value** | Define el valor inicial o contenido de un campo de entrada. |
| **name** | Nombre del elemento usado para identificar los datos al enviar un formulario. |
| **required** | Indica que un campo de formulario es obligatorio. |
| **checked** | Marca un elemento como seleccionado por defecto en `<input>` (ej: `radio`, `checkbox`). |
| **selected** | Marca un elemento como seleccionado por defecto en `<option>` o `<select>`. |
| **multiple** | Permite seleccionar múltiples opciones en `<select>`. |
| **hidden** | Oculta un elemento para que no sea visible en la página. |
| **disabled** | Deshabilita un elemento para que el usuario no pueda interactuar con él. |
| **width / height** | Establecen el ancho y el alto de elementos como `<img>` o `<canvas>`. |


````python
%%HTML
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Demostración de Atributos HTML</title>
    <style>
        /* Estilos básicos para la demostración de 'class' */
        .caja-resaltada {
            background-color: #f0f8ff;
            border: 2px solid #007bff;
            padding: 15px;
            margin-bottom: 20px;
            border-radius: 5px;
        }
    </style>
</head>
<body>

    <h1>Ejemplos de Atributos en HTML</h1>

    <hr>

    <h2>1. Atributos Globales y de Estilo</h2>
    
    <div id="seccion-unica" class="caja-resaltada" style="color: #333;" title="¡Hola! Soy un tooltip creado por el atributo 'title'">
        <p>
            Este bloque `div` utiliza:
            <ul>
                <li><strong>id="seccion-unica"</strong>: Para identificarlo de forma única.</li>
                <li><strong>class="caja-resaltada"</strong>: Para aplicarle los estilos CSS definidos en la cabecera.</li>
                <li><strong>style="color: #333;"</strong>: Para cambiar el color del texto directamente (inline).</li>
                <li><strong>title="..."</strong>: Pasa el cursor del ratón sobre este recuadro y verás un mensaje flotante.</li>
            </ul>
        </p>
    </div>

    <p hidden>
        Este párrafo tiene el atributo <strong>hidden</strong>. Si estás leyendo esto en el código, ¡felicidades! Pero en el navegador web no debería verse.
    </p>

    <hr>

    <h2>2. Enlaces y Multimedia</h2>

    <p>
        Visita <a href="https://es.wikipedia.org" target="_blank">Wikipedia</a>. <br>
        (Este enlace usa <strong>href</strong> para definir a dónde ir y <strong>target="_blank"</strong> para abrirse en una pestaña nueva).
    </p>

    <p>Aquí tienes un ejemplo de una imagen:</p>
    <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/6/61/HTML5_logo_and_wordmark.svg/150px-HTML5_logo_and_wordmark.svg.png" 
         alt="Logotipo oficial de HTML5" 
         width="150" 
         height="150">
    <p>
        <small>La imagen usa <strong>src</strong> para cargar el archivo, <strong>alt</strong> por si falla la carga o para lectores de pantalla, y <strong>width/height</strong> para definir su tamaño a 150x150 píxeles.</small>
    </p>

    <hr>

    <h2>3. Formularios e Inputs</h2>

    <form action="#" method="get" class="caja-resaltada">
        
        <label for="nombreUsuario">Nombre de usuario:</label><br>
        <input type="text" 
               id="nombreUsuario" 
               name="usuario" 
               placeholder="Escribe tu usuario aquí" 
               value="Invitado" 
               required>
        <p><small>Este campo usa <strong>type="text"</strong>, un <strong>placeholder</strong> de sugerencia, un <strong>value</strong> inicial ("Invitado"), un <strong>name</strong> para el envío de datos, y es <strong>required</strong> (obligatorio).</small></p>

        <label for="clave">Contraseña (Deshabilitada):</label><br>
        <input type="password" 
               id="clave" 
               name="contrasena" 
               disabled>
        <p><small>Este campo oculta el texto con <strong>type="password"</strong>, pero tiene el atributo <strong>disabled</strong>, por lo que no puedes escribir en él.</small></p>

        <label>
            <input type="checkbox" name="boletin" value="suscrito" checked>
            Quiero recibir el boletín semanal.
        </label>
        <p><small>Es un <strong>type="checkbox"</strong> que aparece marcado por defecto gracias al atributo <strong>checked</strong>.</small></p>

        <label for="lenguajes">Selecciona tus lenguajes favoritos (puedes elegir varios):</label><br>
        <select name="lenguajes" id="lenguajes" multiple size="4">
            <option value="html" selected>HTML (Seleccionado por defecto)</option>
            <option value="css">CSS</option>
            <option value="js">JavaScript</option>
            <option value="python">Python</option>
        </select>
        <p><small>Este menú desplegable usa <strong>multiple</strong> para permitir varias selecciones, y la opción de HTML usa <strong>selected</strong> para estar resaltada desde el principio.</small></p>

        <button type="submit">Enviar Formulario</button>
    </form>

</body>
</html>
````

## La Estructura Estándar

Todo archivo HTML profesional sigue una estructura básica. Pensemos en esto como los cimientos obligatorios de cualquier edificio.

````python
%%HTML
<!-- Ejemplo de estructura básica HTML -->
<!DOCTYPE html>
<html>
  <head>
    <title>Título de la Página</title>
    <meta charset="UTF-8">
    <meta name="description" content="Descripción de la página"><!-- Descripcion de la pagina -->
    <meta name="keywords" content="palabras clave"><!-- Palabras clave -->
    <meta name="author" content="Nombre del Autor"><!-- Autor -->
    <meta name="viewport" content="width=device-width, initial-scale=1.0">  <!-- Tamaño de la pagina -->
    <meta name="theme-color" content="#000000"><!-- Color de la pagina -->
    <meta name="robots" content="index, follow"><!-- robots  -->
    <link rel="icon" href="favicon.ico">
    <link rel="stylesheet" href="styles.css">
    <!--open graph-->
    <meta property="og:title" content="Título de la Página">
    <meta property="og:description" content="Descripción de la página">
    <meta property="og:image" content="imagen.jpg">
    <meta property="og:url" content="https://www.ezequiel.com">
    <meta property="og:type" content="website">
    
    
  </head>
  <body>
    <!-- Todo lo que el usuario ve va aquí -->
    <h1>Hola Mundo!</h1>
  </body>
</html>
````

### Explicación de las etiquetas en el `<head>`:

*   **`<title>`**: Define el título que aparece en la pestaña del navegador y es fundamental para el SEO (posicionamiento en buscadores).
*   **`<meta charset="UTF-8">`**: Establece la codificación de caracteres universal, permitiendo mostrar correctamente tildes, eñes y símbolos especiales.
*   **`<meta name="description">`**: Proporciona un resumen del contenido de la página que los buscadores muestran en sus resultados.
*   **`<meta name="keywords">`**: Lista de palabras clave relevantes para el sitio (aunque actualmente tiene menos peso en el SEO moderno).
*   **`<meta name="author">`**: Especifica el nombre del creador o autor del documento.
*   **`<meta name="viewport">`**: Configura la vista en dispositivos móviles, asegurando que el contenido se adapte al ancho de la pantalla y tenga el zoom correcto.
*   **`<meta name="theme-color">`**: Define el color de la interfaz del navegador (como la barra de direcciones) en dispositivos móviles compatibles.
*   **`<meta name="robots">`**: Indica a los motores de búsqueda si deben indexar la página (`index`) y si deben seguir los enlaces que contiene (`follow`).
*   **`<link rel="icon">`**: Enlaza el archivo de imagen que se mostrará como icono (favicon) en la pestaña del navegador.
*   **`<link rel="stylesheet">`**: Vincula una hoja de estilos CSS externa para aplicar el diseño visual al documento HTML.


### Desglose:
1. `<!DOCTYPE html>`: Le dice al navegador "Hey, esto es un documento HTML5 moderno".
2. `<html>`: Es la raíz de todo. Todo nuestro código vive aquí dentro.
3. `<head>`: Es el "cerebro" de la página. Aquí va información que el usuario no ve directamente en el contenido, pero es importante (como el título en la pestaña del navegador o la configuración de caracteres).
4. `<body>`: Es el "cuerpo". **Todo** lo que queremos mostrar en la pantalla (textos, imágenes, botones) va aquí.

### 1. Texto
Para organizar la lectura, usamos encabezados y párrafos.

- `<h1>` hasta `<h6>`: Encabezados. `<h1>` es el más importante (el título principal), `<h6>` el menos importante.
- `<p>`: Párrafo normal de texto.
- `<strong>`: Texto en **negrita** (indica importancia).
- `<em>`: Texto en *cursiva* (indica énfasis).

````python
%%HTML
<!-- Ejemplo de texto -->
<h1>Bienvenidos a mi Blog</h1>
<h2>Capítulo 1: El comienzo</h2>
<p>Este es un párrafo de introducción. Aquí explico cosas importantes.</p>
<p>Este es otro párrafo con una palabra en <strong>negrita</strong> y otra en <em>cursiva</em>.</p>
````

### 2. Listas
Hay dos tipos principales:
- **No ordenadas** (`<ul>`): Usan viñetas (puntitos). Común para menús o listas de compras.
- **Ordenadas** (`<ol>`): Usan números (1, 2, 3...). Común para pasos de una receta.
- `<li>`: Cada elemento de la lista (List Item).

````python
%%HTML
<!-- Lista No Ordenada -->
<ul>
  <li>Manzanas</li>
  <li>Peras</li>
  <li>Bananas</li>
</ul>

<!-- Lista Ordenada -->
<ol>
  <li>Precalentar el horno</li>
  <li>Mezclar los ingredientes</li>
  <li>Hornear por 30 min</li>
</ol>
````

### 3. Enlaces e Imágenes
El poder de la web está en conectar cosas.

- `<a>` (Anchor): Crea un enlace. Necesita el atributo `href` para saber a dónde ir.
- `<img>`: Muestra una imagen. Es una etiqueta especial que **no se cierra**. Necesita `src` (fuente) y `alt` (texto alternativo si la imagen falla).

````python
%%HTML
<!-- Enlace a Google -->
<a href="https://www.google.com" target="_blank">Ir a Google</a>

<!-- Imagen -->
<img src="https://www.google.com/imgres?q=ipm%20&imgurl=https%3A%2F%2Fipm.edu.ar%2Fassets%2Fimg%2Flogo.png&imgrefurl=https%3A%2F%2Fwww.ipm.edu.ar%2F&docid=gwswGt7yH6hzSM&tbnid=emVZuQL7MuASxM&vet=12ahUKEwi_0JLk69OSAxVHqJUCHaY5MvYQnPAOegQIGRAB..i&w=267&h=333&hcb=2&ved=2ahUKEwi_0JLk69OSAxVHqJUCHaY5MvYQnPAOegQIGRAB" alt="Una imagen de ejemplo" width="150">
````

### 4. Formularios e Inputs
Para interactuar con el usuario (recibir datos), usamos formularios. La etiqueta estrella es `<input>`.

- `<input>`: Es un **elemento de reemplazo** (vacío), no tiene etiqueta de cierre. Su apariencia y comportamiento cambian drásticamente según el atributo `type`.
- `<label>`: Define una etiqueta para un elemento de formulario. Mejora la accesibilidad y permite que al hacer clic en el texto se active el campo asociado (usando el atributo `for` vinculado al `id` del input).
- `<button>`: Define un botón clickable. A diferencia de los inputs, puede contener contenido complejo como texto e imágenes dentro de sus etiquetas de apertura y cierre.


````python
%%HTML
<!-- Campo de texto normal -->
 <main>
<label>Nombre:</label>
<input type="text" placeholder="Escribe tu nombre" required>
<br>

<!-- Campo de contraseña (los caracteres se ocultan) -->
<label>Contraseña:</label>
<input type="password" placeholder="Tu clave secreta">
<br>

<!-- Campo numérico -->
<label>Edad:</label>
<input type="number" min="18" max="99">
<br>

<!-- Casilla de verificación -->
<label>Acepto términos:</label>
<input type="checkbox" name="terms">
<br>

<!-- Selector de fecha -->
<label>Fecha de Nacimiento:</label>
<input type="date">
<br>

<!-- Botón de envío -->
<button type="submit">Enviar Datos</button>
<br>
<a href="index.html">
    <button type="button">Volver al Index</button>
</a>
</main>
````

### 5. Etiquetas Genéricas (No Semánticas)

A veces necesitamos agrupar cosas solo por diseño o estructura, sin que signifique nada especial para el navegador. Para eso existen `<div>` y `<span>`.

- **`<div>` (Division)**: Es un contenedor de **bloque**. Se usa para agrupar grandes secciones o cajas. Es como una caja de cartón genérica.
- **`<span>`**: Es un contenedor de **línea**. Se usa para seleccionar una pequeña parte de texto dentro de una línea.

**Diferencia Clave**:
- `<div>` rompe la línea (cae abajo).
- `<span>` se mantiene en la misma línea.

*Nota: Abusar de estas etiquetas se llama "divitis". Siempre prefiere usar etiquetas semánticas (que veremos abajo) si es posible.*

````python
%%HTML
<!-- Ejemplo de DIV (Caja grande) -->
<div style="color:black; background-color: lightgray; padding: 10px;">
  <h3>Esta es una caja genérica</h3>
  <p>El div agrupa este título y este párrafo.</p>
</div>

<!-- Ejemplo de SPAN (Detalle pequeño) -->
<p>
  Este es un texto normal, pero <span style="color: red; font-weight: bold;">esta parte resalta</span> gracias al span.
</p>
````

## Introducción a la Semántica

Antes, los desarrolladores usaban una etiqueta genérica llamada `<div>` para todo. Pero eso es como guardar todas tus cosas en cajas idénticas sin etiquetas: un caos para encontrar algo.

HMTL5 introdujo **etiquetas semánticas**. Estas etiquetas tienen un nombre que explica **qué son**.

- `<header>`: La cabecera de la página (logo, menú principal).
- `<nav>`: Bloque de enlaces de navegación.
- `<main>`: El contenido principal y único de esa página.
- `<section>`: Una sección temática dentro del contenido.
- `<aside>`: Contenido relacionado de forma indirecta (barras laterales, notas al margen).
- `<article>`: Un contenido independiente (como un post de un blog o una noticia).
- `<footer>`: El pie de página (copyright, contacto).

Usar estas etiquetas ayuda a los buscadores (como Google) a entender tu página y a las personas que usan lectores de pantalla.

````python
%%HTML
<header>
    <h1>Mi Sitio Web Semántico</h1>
    <nav>
        <ul>
            <li><a href="#">Inicio</a></li>
            <li><a href="#">Servicios</a></li>
            <li><a href="#">Contacto</a></li>
        </ul>
    </nav>
</header>

<main>
    <section>
        <h2>Sección Principal</h2>
        <p>Este es el contenido principal de la página utilizando etiquetas semánticas de HTML5.</p>
        
        <article>
            <h3>Artículo Informativo</h3>
            <p>Contenido detallado dentro de un artículo independiente.</p>
        </article>
    </section>

    <aside>
        <h4>Barra Lateral</h4>
        <p>Información relacionada o enlaces de interés.</p>
    </aside>
</main>

<footer>
    <p>&copy; 2023 Mi Proyecto de Python - Clase OC</p>
</footer>
````
